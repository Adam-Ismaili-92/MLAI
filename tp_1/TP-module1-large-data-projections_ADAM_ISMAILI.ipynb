# Multidimensional projections in Python
Most Python libraries follow the conventions of scikit-learn and provide an API compatible with scikit-learn.
The simplest projection is [PCA](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html).

# Iris Dataset
You can check what is the Iris Dataset in here: https://en.wikipedia.org/wiki/Iris_flower_data_set

First, show the dataset with the first two dimensions

```python
%matplotlib inline

import matplotlib.pyplot as plt

# unused but required import for doing 3d projections with matplotlib < 3.2
import mpl_toolkits.mplot3d  # noqa: F401

from sklearn import datasets
from sklearn.decomposition import PCA

# import some data to play with
iris = datasets.load_iris()
X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

# PCA
We apply PCA and show the first two principal components

```python
# To getter a better understanding of interaction of the dimensions
# plot the first three PCA dimensions
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)

ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
#ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
#ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
#ax.w_zaxis.set_ticklabels([])

plt.show()
```

# More projections
Let's try the MNIST dataset with MDS and t-SNE

```python
from sklearn.manifold import (MDS, TSNE)
import pandas as pd
import holoviews as hv
hv.extension("bokeh")
```

# MDS

```python
iris_mds = MDS(n_components=2, n_init=1, max_iter=100)
data = X.copy()
projection_mds = iris_mds.fit_transform(data, y)
iris_mds_df = pd.DataFrame(data={'Sepal length': projection_mds[:, 0], 'Sepal width': projection_mds[:, 1], 'cat': y})
iris_mds_df['cat'] = iris_mds_df['cat'].astype("string")
points = hv.Points(iris_mds_df, kdims=['Sepal length', 'Sepal width'], vdims=['cat'])
points.opts(color='cat', cmap='Set1')
```

```python
iris_tsne = TSNE(n_components=2, init="pca", random_state=0)
data = X.copy()
projection_tsne = iris_tsne.fit_transform(data, y)
iris_tsne_df = pd.DataFrame(data={'Sepal length': projection_tsne[:, 0], 'Sepal width': projection_tsne[:, 1], 'cat': y})
iris_tsne_df['cat'] = iris_tsne_df['cat'].astype("str")
points = hv.Points(iris_tsne_df, kdims=['Sepal length', 'Sepal width'], vdims=['cat'])
points.opts(color='cat', cmap='Set1')
```

# Scaling UMAP
UMAP is superior to t-SNE in that it has a global scale optimization as well as a local-scale optimization.
Therefore, for a large dataset, it can learn the manifold on a portion of the data and project the remaining part of the data very quickly.

```python
from sklearn.model_selection import train_test_split

# Split the digit dataset into a train set and a test set
X_train, X_test, y_train, y_test = train_test_split(digits.data,
                                                    digits.target,
                                                    stratify=digits.target,
                                                    random_state=42)
```

We will use UMAP on the train dataset

```python
trans = umap.UMAP(n_neighbors=5, random_state=42).fit(X_train)
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 8]
%matplotlib inline
plt.scatter(trans.embedding_[:, 0], trans.embedding_[:, 1], s= 5, c=y_train, cmap='Spectral')
plt.title('Embedding of the training set by UMAP', fontsize=24)
```

Use the trained model to transform the remaining part of the dataset, the test part:

```python
%time test_embedding = trans.transform(X_test)
```

```python
plt.scatter(test_embedding[:, 0], test_embedding[:, 1], s= 5, c=y_test, cmap='Spectral')
plt.title('Embedding of the test set by UMAP', fontsize=24);
```

Swiss Roll And Swiss-Hole Reduction

The code at "https://scikit-learn.org/stable/auto_examples/manifold/plot_swissroll.html#sphx-glr-auto-examples-manifold-plot-swissroll-py" seeks to compare two popular non-linear dimensionality techniques, T-distributed Stochastic Neighbor Embedding (t-SNE) and Locally Linear Embedding (LLE), on the classic Swiss Roll dataset.

First, reproduce this exercice here.
Second, add UMAP and compare the results with t-SNE and LLE.

```python
sr_points, sr_color = datasets.make_swiss_roll(n_samples=1500, random_state=0)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
fig.add_axes(ax)
ax.scatter(
    sr_points[:, 0], sr_points[:, 1], sr_points[:, 2], c=sr_color, s=50, alpha=0.8
)
ax.set_title("Swiss Roll in Ambient Space")
ax.view_init(azim=-66, elev=12)
_ = ax.text2D(0.8, 0.05, s="n_samples=1500", transform=ax.transAxes)
```

```python
from sklearn import manifold

sr_lle, sr_err = manifold.locally_linear_embedding(
    sr_points, n_neighbors=12, n_components=2
)

sr_tsne = manifold.TSNE(n_components=2, perplexity=40, random_state=0).fit_transform(
    sr_points
)

sr_umap = umap.UMAP(n_components=2).fit_transform(
    sr_points
)

fig, axs = plt.subplots(figsize=(8, 12), nrows=3)
axs[0].scatter(sr_lle[:, 0], sr_lle[:, 1], c=sr_color)
axs[0].set_title("LLE Embedding of Swiss Roll")
axs[1].scatter(sr_tsne[:, 0], sr_tsne[:, 1], c=sr_color)
axs[1].set_title("t-SNE Embedding of Swiss Roll")
axs[2].scatter(sr_umap[:, 0], sr_umap[:, 1], c=sr_color)
_ = axs[2].set_title("UMAP Embedding of Swiss Roll")
```

```python
sh_points, sh_color = datasets.make_swiss_roll(
    n_samples=1500, hole=True, random_state=0
)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")
fig.add_axes(ax)
ax.scatter(
    sh_points[:, 0], sh_points[:, 1], sh_points[:, 2], c=sh_color, s=50, alpha=0.8
)
ax.set_title("Swiss-Hole in Ambient Space")
ax.view_init(azim=-66, elev=12)
_ = ax.text2D(0.8, 0.05, s="n_samples=1500", transform=ax.transAxes)
```

```python
sh_lle, sh_err = manifold.locally_linear_embedding(
    sh_points, n_neighbors=12, n_components=2
)

sh_tsne = manifold.TSNE(
    n_components=2, perplexity=40, init="random", random_state=0
).fit_transform(
    sh_points
)

sh_umap = umap.UMAP(n_components=2).fit_transform(
    sh_points
)

fig, axs = plt.subplots(figsize=(8, 12), nrows=3)
axs[0].scatter(sh_lle[:, 0], sh_lle[:, 1], c=sh_color)
axs[0].set_title("LLE Embedding of Swiss-Hole")
axs[1].scatter(sh_tsne[:, 0], sh_tsne[:, 1], c=sh_color)
axs[1].set_title("t-SNE Embedding of Swiss-Hole")
axs[2].scatter(sh_umap[:, 0], sh_umap[:, 1], c=sh_color)
_ = axs[2].set_title("UMAP Embedding of Swiss-Hole")
```

LLE very capably unrolls the data and even preserves the hole. t-SNE and UMAP seem to clump sections of points together, but, they preserve the general topology of the original data, and UMAP is doing it better than t-SNE by preserving the winding appearance of the data.

The S-curve dataset

The code at "https://scikit-learn.org/stable/auto_examples/manifold/plot_compare_methods.html#sphx-glr-auto-examples-manifold-plot-compare-methods-py" present examples of manifold learning on the S-curve dataset.

Create a code that compares MDS, t-SNE and UMAP on the S-curve dataset.

```python
%matplotlib inline

import matplotlib.pyplot as plt

# unused but required import for doing 3d projections with matplotlib < 3.2
import mpl_toolkits.mplot3d  # noqa: F401
from matplotlib import ticker

from sklearn import datasets, manifold

n_samples = 1500
S_points, S_color = datasets.make_s_curve(n_samples, random_state=0)
```

```python
def plot_3d(points, points_color, title):
    x, y, z = points.T

    fig, ax = plt.subplots(
        figsize=(6, 6),
        facecolor="white",
        tight_layout=True,
        subplot_kw={"projection": "3d"},
    )
    fig.suptitle(title, size=16)
    col = ax.scatter(x, y, z, c=points_color, s=50, alpha=0.8)
    ax.view_init(azim=-60, elev=9)
    ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
    ax.zaxis.set_major_locator(ticker.MultipleLocator(1))

    fig.colorbar(col, ax=ax, orientation="horizontal", shrink=0.6, aspect=60, pad=0.01)
    plt.show()


def plot_2d(points, points_color, title):
    fig, ax = plt.subplots(figsize=(3, 3), facecolor="white", constrained_layout=True)
    fig.suptitle(title, size=16)
    add_2d_scatter(ax, points, points_color)
    plt.show()


def add_2d_scatter(ax, points, points_color, title=None):
    x, y = points.T
    ax.scatter(x, y, c=points_color, s=50, alpha=0.8)
    ax.set_title(title)
    ax.xaxis.set_major_formatter(ticker.NullFormatter())
    ax.yaxis.set_major_formatter(ticker.NullFormatter())


plot_3d(S_points, S_color, "Original S-curve samples")
```

```python
md_scaling = manifold.MDS(
    n_components=2,
    max_iter=50,
    n_init=4,
    random_state=0,
    normalized_stress=False,
)
S_scaling = md_scaling.fit_transform(S_points)

plot_2d(S_scaling, S_color, "MDS")

t_sne = manifold.TSNE(
    n_components=2,
    perplexity=30,
    init="random",
    n_iter=250,
    random_state=0,
)
S_t_sne = t_sne.fit_transform(S_points)

plot_2d(S_t_sne, S_color, "TSNE")

u_map = umap.UMAP(n_components=2)
S_umap = u_map.fit_transform(S_points)

plot_2d(S_umap, S_color, "UMAP")
```

MDS seems to greatly preserve the shape of the data, but UMAP and TSNE seem to better display the data topology.
